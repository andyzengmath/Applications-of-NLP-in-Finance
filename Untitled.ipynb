```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
from scipy.stats import skew, norm, probplot
import time
from sklearn.preprocessing import OneHotEncoder, RobustScaler, PolynomialFeatures,StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score,mean_squared_log_error
from sklearn.feature_selection import mutual_info_regression
from sklearn.linear_model import Ridge, HuberRegressor, LinearRegression,Lasso,Ridge
from sklearn.svm import SVR
from sklearn.cluster import KMeans
import catboost as cb
from xgboost import XGBRegressor
from mlxtend.regressor import StackingCVRegressor
from sklearn.svm import SVC
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor, StackingRegressor, ExtraTreesRegressor
import catboost as cb
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
y=train['SalePrice']
df=train.drop(['SalePrice'],axis=1)
df.set_index('Id')
test.set_index('Id')
```

```python
y=np.log(y)
max(y)
```

```python
df.columns
```

```python
null_list_q=[]
null_list_c=[]
for col in df.columns:
    null=df[col].isnull().sum()
    test_null=test[col].isnull().sum()
    if null+test_null!=0:
        if df[col].dtype=='object':
            null_list_c.append([col,null,test_null,null+test_null])
        else:
            null_list_q.append([col,null,test_null,null+test_null])
null_df_q = pd.DataFrame(null_list_q,columns=['feature','train_null','test_null','total_null'])
null_df_q.set_index('feature')
null_df_c = pd.DataFrame(null_list_c,columns=['feature','train_null','test_null','total_null'])
null_df_c.set_index('feature')
```

```python
missing_q=null_df_q.sort_values(by='total_null',ascending= False)
missing_c=null_df_c.sort_values(by='total_null',ascending= False)
```

```python
missing_q
```

```python
test.shape
```

```python
missing_c
```

```python
for feature in missing_c['feature']:
    df[feature]=df[feature].fillna('No')
    test[feature]=test[feature].fillna('No')
```

```python
for feature in missing_q['feature']:
    df[feature]=df[feature].fillna(df[feature].median())
    test[feature]=test[feature].fillna(test[feature].median())
```

```python
feature_c=[]
feature_q=[]
for col in df.columns:
    if df[col].dtype=='object':
        feature_c.append(col)
    else:
        feature_q.append(col)
```

```python
enc=OneHotEncoder(sparse=False)
enc_cols=pd.DataFrame(enc.fit_transform(df[feature_c]))
enc_cols.index=df.index
enc_cols.columns=enc.get_feature_names(feature_c)
df=df.drop(feature_c,axis=1)
df=pd.concat([df,enc_cols],axis=1)
```

```python
enc=OneHotEncoder(sparse=False)
enc_cols_test=pd.DataFrame(enc.fit_transform(test[feature_c]))
enc_cols_test.index=test.index
enc_cols_test.columns=enc.get_feature_names(feature_c)
test=test.drop(feature_c,axis=1)
test=pd.concat([test,enc_cols_test],axis=1)
```

```python
df, test = df.align(test, join='inner', axis=1)  # inner join
```

```python
scaler=StandardScaler()
df_scaled = scaler.fit_transform(df[feature_q])
df[feature_q]=df_scaled
```

```python
scaler=StandardScaler()
test_scaled = scaler.fit_transform(test[feature_q])
test[feature_q]=test_scaled
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,y,test_size=0.2,random_state=1)
```

```python
col=[]
for i in df.columns:
    col.append(i)
```

```python
ranking=[]
```

```python
lr=LinearRegression()
```

```python
for i in range(df.shape[1]):
    perf=[]
    if i%10==0:
        print(i)
    for j in range(df.shape[1]-i):
        feature_try=col[j]
        ranking_copy=ranking.copy()
        ranking_copy.append(feature_try)
        X_train_try=X_train[ranking_copy]
        lr_model_try=lr.fit(X_train_try,y_train)
        y_pred=lr_model_try.predict(X_test[ranking_copy])
        score=mean_squared_error(y_pred,y_test)
        perf.append(score)
    min_perf=min(perf)
    winner_index=perf.index(min_perf)
    winner=col[winner_index]
    ranking.append(winner)
    col.remove(winner)
```

```python
ranking_perf=[]
for i in range(df.shape[1]):
    X_train_rank=X_train[ranking[:i+1]]
    Lr_model_rank=lr.fit(X_train_rank,y_train)
    y_pred=Lr_model_rank.predict(X_test[ranking[:i+1]])
    score_rank=mean_squared_error(y_pred,y_test)
    ranking_perf.append(score_rank)
    if i%20==0:
        print(i)
```

```python
ranking
```

```python
number=ranking_perf.index(min(ranking_perf))
number
```

```python
features=ranking[:number+1]
```

```python
X_modeling=df[features]
y_modeling=y
test_modeling=test[features]
```

```python
final_model=lr.fit(X_modeling,y_modeling)
final_prediction=final_model.predict(test_modeling)
final_prediction_exp=np.exp(final_prediction)
```

```python
c=final_prediction_exp
```

```python
submission = pd.DataFrame(c, index = test.index+1461)
submission
```
